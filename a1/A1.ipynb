```python
# 24. Do the following pre- processing techniques on the different datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load datasets
student_df = pd.read_csv("student.csv")
airlines_df = pd.read_csv("airlines_flights_data.csv")
flipkart_df = pd.read_csv("Flipkart-Laptops.csv")
```

```python
student_df.head()
```

```python
airlines_df.head()
```

```python
flipkart_df.head()
```

```python
# cleaning the data
# ---- Student ----
# Fill missing values with mean for numeric, mode for categorical
student_df = student_df.fillna({
    col: student_df[col].mean() if student_df[col].dtype != 'object' else student_df[col].mode()[0]
    for col in student_df.columns
})
# Drop duplicates
student_df = student_df.drop_duplicates()

# ---- Airlines ----
airlines_df = airlines_df.drop_duplicates()
airlines_df = airlines_df.fillna(method='ffill')  # forward fill

# ---- Flipkart ----
flipkart_df = flipkart_df.drop_duplicates()
flipkart_df = flipkart_df.fillna('Unknown')  # replace missing text with 'Unknown'
```

```python
# checking for outliers and removing them
def remove_outliers(df, cols):
    for col in cols:
        if df[col].dtype != 'object':
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR
            df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

student_df = remove_outliers(student_df, student_df.select_dtypes(include=np.number).columns)
airlines_df = remove_outliers(airlines_df, airlines_df.select_dtypes(include=np.number).columns)
flipkart_df = remove_outliers(flipkart_df, flipkart_df.select_dtypes(include=np.number).columns)
```

```python
# Handling Categorical Data
# Convert categorical columns using one-hot encoding
student_encoded = pd.get_dummies(student_df, drop_first=True)
airlines_encoded = pd.get_dummies(airlines_df, drop_first=True)
flipkart_encoded = pd.get_dummies(flipkart_df, drop_first=True)
```

```python
student_encoded.head()
```

```python
airlines_encoded.head()
```

```python
flipkart_encoded.head()
```

```python
# Scaling & Normalization
scaler_std = StandardScaler()
scaler_minmax = MinMaxScaler()

# Apply only on numeric columns
student_scaled = student_encoded.copy()
student_scaled[student_encoded.select_dtypes(include=np.number).columns] = scaler_std.fit_transform(student_encoded.select_dtypes(include=np.number))
```

```python
student_scaled.head()
```

```python
airlines_scaled = airlines_encoded.copy()
airlines_scaled[airlines_encoded.select_dtypes(include=np.number).columns] = scaler_std.fit_transform(airlines_encoded.select_dtypes(include=np.number))
```

```python

flipkart_scaled = flipkart_encoded.copy()
flipkart_scaled[flipkart_encoded.select_dtypes(include=np.number).columns] = scaler_minmax.fit_transform(flipkart_encoded.select_dtypes(include=np.number))
flipkart_scaled.head()
```

## Unit 2 (Graphical Visualization) 

```python
import matplotlib.pyplot as plt

student_df = pd.read_csv("student.csv")

# Bar chart: Average percentage by subject
student_df.groupby('Subject')['Percentage'].mean().plot(kind='bar')
plt.title("Average Percentage by Subject")
plt.show()

# Pie chart: Count of students per subject
student_df['Subject'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Student Distribution by Subject")
plt.show()

# Histogram: Percentage distribution
student_df['Percentage'].hist(bins=10)
plt.title("Percentage Distribution")
plt.show()

# Line chart: Compare Percentage vs Past
plt.plot(student_df['Percentage'], label='Current')
plt.plot(student_df['Percentage_Past'], label='Past')
plt.legend()
plt.title("Current vs Past Performance")
plt.show()

# Dot Plot: Age vs Percentage
plt.scatter(student_df['Age'], student_df['Percentage'])
plt.xlabel("Age")
plt.ylabel("Percentage")
plt.title("Age vs Percentage")
plt.show()

# Ogive
student_df['Percentage'].hist(bins=10, cumulative=True, histtype='step')
plt.title("Ogive - Cumulative Percentage Distribution")
plt.show()

# Descriptive stats
student_df[['Age','Percentage','Percentage_Past']].describe()
```

```python
airlines_df = pd.read_csv("airlines_flights_data.csv")

# Bar chart: Avg price per airline
airlines_df.groupby('airline')['price'].mean().plot(kind='bar', figsize=(10,5))
plt.title("Average Price per Airline")
plt.show()

# Pie chart: Flight class distribution
airlines_df['class'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Class Distribution")
plt.show()

# Scatter plot: Duration vs Price
plt.scatter(airlines_df['duration'], airlines_df['price'], alpha=0.5)
plt.xlabel("Duration (hrs)")
plt.ylabel("Price")
plt.title("Duration vs Price")
plt.show()

# Line chart: Avg price vs days left
airlines_df.groupby('days_left')['price'].mean().plot(kind='line')
plt.title("Price Trend by Days Left")
plt.show()

# Histogram: Price distribution
airlines_df['price'].hist(bins=30)
plt.title("Price Distribution")
plt.show()

# Ogive
airlines_df['price'].hist(bins=30, cumulative=True, histtype='step')
plt.title("Ogive of Flight Prices")
plt.show()

# Dot plot: Stops vs Price
plt.scatter(airlines_df['stops'], airlines_df['price'], alpha=0.5)
plt.title("Stops vs Price")
plt.xlabel("Stops")
plt.ylabel("Price")
plt.show()

# Descriptive stats
airlines_df[['duration','days_left','price']].describe()
```

```python
flipkart_df = pd.read_csv("Flipkart-Laptops.csv")

def clean_price(series):
    return (
        series.replace('[₹, ]', '', regex=True)   # remove ₹, commas, spaces
              .replace(['NIL', ''], np.nan)       # replace NIL and empty string with NaN
              .astype(float)                      # convert to float
    )

flipkart_df['Actual price'] = clean_price(flipkart_df['Actual price'])
flipkart_df['Discount price'] = clean_price(flipkart_df['Discount price'])

# Extract brand (first word of product name)
flipkart_df['Company'] = flipkart_df['Product Name'].str.split().str[0]

# Bar chart: Count of laptops by company
flipkart_df['Company'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Number of Laptops by Company")
plt.show()

# Pie chart: Top 5 companies
flipkart_df['Company'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Laptop Brands")
plt.show()

# Histogram: Actual Price
flipkart_df['Actual price'].astype(float).hist(bins=20)
plt.title("Laptop Price Distribution")
plt.show()

# Scatter plot: Rating vs Price
plt.scatter(flipkart_df['Rating'], flipkart_df['Actual price'].astype(float), alpha=0.5)
plt.xlabel("Rating")
plt.ylabel("Price")
plt.title("Rating vs Price")
plt.show()

# Line chart: Compare Actual vs Discount
flipkart_df[['Actual price','Discount price']].astype(float).head(20).plot(kind='line')
plt.title("Price Comparison (First 20 products)")
plt.show()

# Ogive: Price
flipkart_df['Actual price'].astype(float).hist(bins=20, cumulative=True, histtype='step')
plt.title("Ogive of Laptop Prices")
plt.show()

# Dot plot: Stars vs Price
plt.scatter(flipkart_df['Stars'], flipkart_df['Actual price'].astype(float), alpha=0.5)
plt.xlabel("Stars")
plt.ylabel("Price")
plt.title("Stars vs Price")
plt.show()

# Descriptive stats
flipkart_df[['Actual price','Discount price','Rating','Reviews']].describe()
```
