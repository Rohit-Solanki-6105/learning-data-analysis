```python
# chnaging country null values with the mode values
country_mode = hotel['country'].value_counts().index[0] #mode
hotel['country'] = hotel['country'].fillna(country_mode)
```

```python
children_mode = hotel['children'].value_counts().index[0] #mode
hotel['children'] = hotel['children'].fillna(children_mode)
```
